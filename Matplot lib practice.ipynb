```python
x = [1,2,3,4]
y = [11,22,33,44]
plt.plot(x , y);
```

```python
# First Method
fig = plt.figure()
ax = fig.add_subplot()
plt.show()
```

```python
# 2nd Method
fig = plt.figure()
ax = fig.add_axes([1,1,1,1])
ax.plot(x,y)
plt.show()
```

```python
# 3rd Method (Recomended)
fig, ax = plt.subplots()
ax.plot(x,y);
```

```python
x = [1,2,3,4]
y = [22,33,44,55]
fig , ax = plt.subplots(figsize = (10,5))
ax.plot(x,y)
ax.set(title ='simple Plt ',xlabel = 'x-axis',ylabel='y-axis')
fig.savefig("plt.png")
```

# Plotting from pandas dataFrame

```python
car_sales = pd.read_csv("21708170-car-sales.csv")
car_sales
```

```python
car_sales["Price"]= car_sales["Price"].str.replace('[\$\,\.]', '')
car_sales
```

```python
type(car_sales["Price"][0])
```

```python
car_sales["Price"] = car_sales["Price"].str[:-2]
```

```python
car_sales
```

```python
car_sales["Sale Date"] = pd.date_range('1/1/2022', periods=len(car_sales))
```

```python
car_sales
```

```python
car_sales["Total Price"] = car_sales["Price"].astype(int).cumsum()
car_sales
```

```python
car_sales.plot(x = "Sale Date", y ="Total Price" );
```

```python
car_sales.plot(x = "Odometer (KM)" , y = "Price", kind = "scatter");
```

```python
car_sales
```

```python
car_sales.plot(x = "Make", y = "Odometer (KM)" , kind = "bar");
```

```python
heart_disease = pd.read_csv("heart-disease.csv")
```

```python
heart_disease.head()
```

```python
heart_disease["age"].plot.hist(bins = 50)
```

```python
heart_disease.plot.hist(figsize =(10 , 30) , subplots=True);
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
```

```python
over_50.head()
```

```python
# pyplot method
over_50.plot(kind = "scatter" , x ="age", y ="chol" , c = "target" );
```

```python
#  oo Methode
fig , ax = plt.subplots(figsize = (10, 6))
over_50.plot(kind = 'scatter',
            x='age',
            y='chol',
             c='target',
            ax=ax);
```

```python
# OO Method from scratch
fig , ax = plt.subplots(figsize = (10,6))
scatter = ax.scatter(
                      x=over_50["age"],
                      y=over_50['chol'],
                      c=over_50['target'])
ax.set(title= 'OO Method',
      xlabel='x-axis',
      ylabel='yaxis');
```
